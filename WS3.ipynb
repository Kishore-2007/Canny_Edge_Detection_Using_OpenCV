```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Read the image
image = cv2.imread('tiger.jpg')
```

```python
# Step 2: Resize the image 
resized = cv2.resize(image, (600, 400))
```

```python
# Step 3: Display the original resised image 
plt.subplot(1, 1, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# Step 4: Convert to grayscale
gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
```

```python
# Step 5: Display the grayscale image
plt.subplot(1, 1, 1)
plt.title('Grayscale Image')
plt.imshow(gray, cmap='gray')
plt.axis('off')
```

```python
# Step 6: Apply Blur to reduce noise

blur = cv2.blur(gray, ksize=(5,5))
```

```python
# Step 7: Display the blurred image
plt.subplot(1, 1, 1)
plt.title('Blurred Image')
plt.imshow(blur, cmap='gray')
plt.axis('off')
```

```python
# Step 8: Apply Canny Edge Detection
edges = cv2.Canny(blur, 50, 50)
```

```python
# Step 9: Display the Edge Detected Image
plt.subplot(1, 1, 1)
plt.title('Canny Edge Detection')
plt.imshow(edges, cmap='gray')
plt.axis('off')
```
